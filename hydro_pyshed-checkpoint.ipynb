# Pysheds for processing Digitial Elevation Models

```python
# Pysheds is an open-source library designed to help with the processing of
# Digital Elevation Models(DEMs), particularly for hydrological analysis.
# Catchemnt delineation to accumulation computations are major use-cases of Pysheds
```

```python
from urllib.request import urlretrieve
from os.path import basename

url =  "https://tubcloud.tu-berlin.de/s/ZPiZjX3Nzc6R6RX/download/output_COP30.tif"
urlretrieve(url, basename(url))
```

```python
from pysheds.grid import Grid
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
import numpy as np
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
fn = "output_COP30.tif"
grid = Grid.from_raster(fn)
dem = grid.read_raster(fn) # digital elevation model
```

```python
grid
```

```python
dem
```

```python
fig = plt.figure()
ax = plt.axes(projection=ccrs.PlateCarree())
plt.imshow(dem, extent=grid.extent, cmap="terrain")
plt.colorbar(label="Elevation (m)", orientation="horizontal")
ax.coastlines()
ax.add_feature(cfeature.BORDERS)
```

```python
# Conditioning the elevation data
# pits and flat lands prevent water from flowing so we need to remove such feature 
# from our data and perform further analysis
pit_filled_dem = grid.fill_pits(dem)
flooded_dem = grid.fill_depressions(pit_filled_dem)
conditioned_dem = grid.resolve_flats(flooded_dem)

```

```python
# find out flow direction of water driven by steepest descent
# Flow dorections are important to capture the topology of the drainage network
# pysheds use D8 routing scheme (ESRI), in this method each cell is routed to
# one of the eight neighboring cells based on direction of steepest descent
flowdir = grid.flowdir(conditioned_dem)

```

```python
np.unique(flowdir)
```

```python
plt.imshow(flowdir, extent=grid.extent, cmap="cividis")
```

```python
# computing flow accumulation
# basically calculating the number of cells from which the accumulation
# is taking place
acc = grid.accumulation(flowdir)
fig, ax = plt.subplots()
im = ax.imshow(
    acc,
    extent=grid.extent,
    cmap="cubehelix",
    norm=LogNorm(1,acc.max()),
    interpolation="bilinear",
)
plt.colorbar(im, ax=ax, label="Upstream Cells", orientation="horizontal")
    
```

```python
norm = LogNorm
```

```python
# computing catchment area
# lets day we have pour point maybe a weir at x,y point
x, y = -7.954, 41.47
x_snap, y_snap = grid.snap_to_mask(acc>5000, (x,y)) # snapping the pour point to high elevation
```

```python
catch = grid.catchment(x=x_snap,y=y_snap, fdir=flowdir, xytype="coordinate")
```

```python
grid.clip_to(catch)
clipped_catch = grid.view(catch)
```

```python
clipped_accumulated_flow = grid.accumulation(flowdir)
```

```python
fig, ax = plt.subplots()
im = ax.imshow(
    np.where(clipped_catch, clipped_accumulated_flow, np.nan),
    extent = grid.extent,
    cmap="cubehelix",
    norm=LogNorm(1,clipped_accumulated_flow.max()),
    interpolation="bilinear",
)
plt.colorbar(
    im,
    ax=ax,
    label="Upstream Cells",
)
plt.scatter([x_snap],[y_snap],c="r",s=50)
    
```
